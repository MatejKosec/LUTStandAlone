```python
class Node():
    loc = -1.0;
    left = -1.0;
    right = -1.0;
    def __init__(self,loc, left, right):
        self.loc = loc
        self.left= left
        self.right=right
        return
    def __repr__(self):
        if self.loc == None:
            print None
        else:
            print self.loc
        if self.left == None:
            print None
        else:
            print self.left            
        if self.right == None:
            print None
        else:
            print self.right

def kdtree(point_list, depth=0):
    k = len(point_list[0])

    # Select axis based on depth so that axis cycles through all valid values
    axis = depth % k
    #print 'axis', axis
    #print 'len', len(point_list)
 
    # Sort point list and choose median as pivot element
    point_list = sp.sort(point_list, axis=axis)
    median = len(point_list)/2 # choose median
 
    # Create node and construct subtrees
    if len(point_list)<3:
        return 
    else:
        return Node(point_list[median,:], kdtree(point_list[:median,:], depth + 1), 
                kdtree(point_list[median + 1:,:], depth + 1))
```
